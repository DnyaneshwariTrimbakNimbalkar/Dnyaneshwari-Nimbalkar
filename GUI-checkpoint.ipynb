```python
### square find application

top = tk.Tk()

top.title(' Square of number')   # for name
top.geometry('700x500') 

data = tk.IntVar()

def square():
    num = data.get()
    result = num * num
    messagebox.showinfo('MESSAGE','Square is %i' %result)

l1 = tk.Label(top, text= 'INPUT').place(x = 50, y = 50)
e1 = tk.Entry(top, textvariable = data).place(x = 100, y = 50)
b1 = tk.Button(top, text = 'Square', command = square, font = ('Bold', 14), bg = 'purple',fg = 'yellow').place(x = 120, y = 100)

top.mainloop()  # manully close window
```

```python
# take 2 numbers, add, sub, divide, mult
top = tk.Tk()

top.title(' Addition of number')   # for name
top.geometry('700x500') 

data1 = tk.IntVar()
data2 = tk.IntVar()


def add():
    num1 = data1.get()
    num2 = data2.get()
    result = num1 + num2
    messagebox.showinfo('MESSAGE','Addition is %i' %result)

def sub():
    num1 = data1.get()
    num2 = data2.get()
    result = num1 - num2
    messagebox.showinfo('MESSAGE','Substraction is %i' %result)
def mul():
    num1 = data1.get()
    num2 = data2.get()
    result = num1 * num2
    messagebox.showinfo('MESSAGE','Multiplication is %i' %result)
    
def div():
    num1 = data1.get()
    num2 = data2.get()
    result = num1 /num2
    messagebox.showinfo('MESSAGE','Division is %i' %result)   #tpu- long form
    
    

l1 = tk.Label(top, text= 'NUM1').place(x = 50, y = 50)
l2 = tk.Label(top, text= 'NUM2').place(x = 50, y = 100)
e1 = tk.Entry(top, textvariable = data1).place(x = 100, y = 50)
e1 = tk.Entry(top, textvariable = data2).place(x = 100, y = 95)
b1 = tk.Button(top, text = 'ADD', command = add, font = ('Bold', 14), bg = 'purple',fg = 'yellow').place(x = 190, y = 200)
b2 = tk.Button(top, text = 'SUB', command = sub, font = ('Bold', 14), bg = 'purple',fg = 'yellow').place(x = 250, y = 200)
b3 = tk.Button(top, text = 'MUL', command = mul, font = ('Bold', 14), bg = 'purple',fg = 'yellow').place(x = 310, y = 200)
b4 = tk.Button(top, text = 'DIV', command = div, font = ('Bold', 14), bg = 'purple',fg = 'yellow').place(x = 370, y = 200)

top.mainloop()  # manully close window
```
